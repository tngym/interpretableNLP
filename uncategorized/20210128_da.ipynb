```
import pandas as pd
import numpy as np

data = pd.read_excel('default of credit card clients.xls', index_col=0, header=1)
```

```
from sklearn.model_selection import train_test_split
data_tr, data_ts = train_test_split(data, test_size=0.2, random_state=42, stratify=data["default payment next month"])
```

```
data_tr.head()
```

```
from sklearn.metrics import roc_auc_score

col_to_ar = {}
for col in data_tr.columns[:-1]:
  auc = roc_auc_score(data_tr.values[:,-1], data_tr[col])
  col_to_ar[col] = 2 * auc - 1
  print(f"{col}: {col_to_ar[col]}")

col_to_ar = pd.DataFrame(pd.Series(col_to_ar), columns=["AR"])
col_to_ar["AR (abs)"] = np.abs(col_to_ar["AR"])
col_to_ar.sort_values(by="AR (abs)", ascending=False)
```

```
from sklearn.ensemble import RandomForestClassifier

hyperparameters = {"n_estimators": 500}
```

```
def augment(data_tr):
  data_tr_default = data_tr[data_tr.iloc[:,-1]==True]
  ls_augmented = []
  for i in range(3):
    for col in ["PAY_0", "PAY_2", "PAY_3"]:
      d = data_tr_default.copy()
      d[col] += i
      d = d[d[col] <= 2]
      # d = d[d[col] <= 8]
      ls_augmented.append(d)
  data_augmented_pos = pd.concat(ls_augmented)

  data_tr_nondefault = data_tr[data_tr.iloc[:,-1]==False]
  ls_augmented = []
  for i in range(3):
      d = data_tr_nondefault.copy()
      d["LIMIT_BAL"] *= (d["LIMIT_BAL"] * (4-i)/5).astype(int).round(-4)
      ls_augmented.append(d)
  data_augmented_neg = pd.concat(ls_augmented)

  return pd.concat([data_tr, data_augmented_pos, data_augmented_neg])
```

```
def run_non_augmented(seed):
  data_tr, data_ts = train_test_split(data, test_size=0.2, random_state=seed, stratify=data["default payment next month"])
  clf = RandomForestClassifier(random_state=seed, **hyperparameters)

  clf.fit(data_tr.values[:, :-1], data_tr.values[:,-1])

  pred_test = clf.predict_proba(data_ts.values[:, :-1])
  auc = roc_auc_score(data_ts.values[:, -1], pred_test[:, 1])

  data_ts_augmented = augment(data_ts)
  pred_test = clf.predict_proba(data_ts_augmented.values[:, :-1])
  auc_augmented = roc_auc_score(data_ts_augmented.values[:, -1], pred_test[:, 1])
  return auc, auc_augmented
```

```
def run_augmented(seed):
  data_tr, data_ts = train_test_split(data, test_size=0.2, random_state=seed, stratify=data["default payment next month"])
  data_tr_augmented = augment(data_tr)
  clf = RandomForestClassifier(random_state=seed, **hyperparameters)

  clf.fit(data_tr_augmented.values[:, :-1], data_tr_augmented.values[:,-1])

  pred_test = clf.predict_proba(data_ts.values[:, :-1])

  auc = roc_auc_score(data_ts.values[:, -1], pred_test[:, 1])
  
  data_ts_augmented = augment(data_ts)
  pred_test = clf.predict_proba(data_ts_augmented.values[:, :-1])
  auc_augmented = roc_auc_score(data_ts_augmented.values[:, -1], pred_test[:, 1])
  return auc,auc_augmented
```

```
ls_aucs_n = []
ls_aucs_n_a = []
ls_aucs_a = []
ls_aucs_a_a = []
for seed in range(3):
  auc_n, auc_n_a = run_non_augmented(seed)
  print(f"Non-Augmented:{auc_n}, {auc_n_a}")
  auc_a, auc_a_a = run_augmented(seed)
  print(f"Augmented    :{auc_a}, {auc_a_a}")

  ls_aucs_n.append(auc_n)
  ls_aucs_n_a.append(auc_n_a)
  ls_aucs_a.append(auc_a)
  ls_aucs_a_a.append(auc_a_a)
```
